**Research Question:** *How does geographic location specifically the urban-rural divide impact broadband accessibility in the United States?*

**Problem statement:** Broadband internet access is a critical driver of economic opportunity, education, healthcare, and civic participation. However, a significant digital divide persists in the United States, particularly between urban and rural regions. Rural communities often face limited broadband infrastructure, lower connection speeds, and fewer provider options, reinforcing existing socioeconomic inequalities. This analysis seeks to examine how geographic location specifically the urban-rural divide impacts broadband accessibility and availability.

**Hypothesis:** Rural areas have significantly lower broadband access rates than urban areas, even after accounting for population density and income levels.

#### **Updating the datatypes to relevant requirements**
---

```python
# List of columns to convert to numeric, excluding identifier columns
cols_to_convert = [col for col in df.columns if col not in ['GEO.id', 'GEO.id2', 'GEO.display-label']]

for col in cols_to_convert:
    # Attempt to convert to numeric, coercing errors to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.info() # Check info again to see updated datatypes
df.head() # Check head again
```
